```python
import keras
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('C:\\Users\\user\\Documents\\HousingPrices.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=['SalePrice'])
Y = df[['SalePrice']]
```

```python
model = keras.models.Sequential()
#the number 8 represents the number of columns we are training while the 1 is the one we are testing against
model.add(keras.layers.Dense(8, activation='relu', input_shape=(8,)))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error', metrics = ['accuracy'])
```

```python
model.fit(X, Y, epochs = 500, callbacks = [keras.callbacks.EarlyStopping(patience=3)])
```

```python
test_data = np.array([2001,	866,	1786,	2,	1,	3,	6,	2008])
print(model.predict(test_data.reshape(1,8), batch_size=1))
```
